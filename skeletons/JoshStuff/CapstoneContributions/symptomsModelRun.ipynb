```python
label_names = ['Tan', 'Sunken', 'Black-Lesions', 'Black-Veins', 'Rotting', 'Brown', 'Wilt', 'Blotchy', 'Healthy', 'Oily', 'Yellow', 'Greasy', 'Black', 'Scorched', 'Purple', 'Brown-Lesions', 'Watery', 'Halo', 'Spots', 'Distortion', 'Sunscald', 'Powdery', 'Fuzzy', 'Elliptical-Lesions', 'Parallel-Lesions', 'Green-Lesions', 'Spores', 'Yellow-Lesions', 'Discolored-Streaks', 'White-Mold', 'Circular', 'Rings', 'Collapsed', 'Shriveled', 'Black-Specks', 'Discolored']
model_path = '/content/drive/My Drive/plantVis4/model.keras'
dataset_path = '/content/drive/My Drive/plantVis4/data.csv'

dataset = pd.read_csv(dataset_path)
model = tf.keras.models.load_model(model_path)

def load_and_preprocess(path):
  img = tf.io.read_file(path)
  img = tf.image.decode_jpeg(img, channels=3)
  img = tf.image.resize(img, (256, 256))
  img = img / 255.0
  return img
```

```python
img_path = "/content/tomatoTestImg.webp"#dataset["image"].iloc[0]
true_label = "Early Blight"#dataset["label"].iloc[0]
img = load_and_preprocess(img_path)
img_batch = tf.expand_dims(img, axis=0)

#run the model on an image
prediction = model.predict(img_batch)
```

```python
#print the image and the predicted label
plt.imshow(np.squeeze(img))
print(prediction)
#plt.xlabel(label_names[ tf.argmax(prediction, axis=1).numpy()[0] ])

output = prediction[0]   # shape (35,)

threshold = 0

for label, value in zip(label_names, output):
    if value > threshold:
        print(f"{label}: {value:.4f}")
```
