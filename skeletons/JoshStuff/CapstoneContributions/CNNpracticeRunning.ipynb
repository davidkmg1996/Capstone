```python
#this whole block is just housekeeping
  #imports
  #associating label names with indexes
  #importing images to run the model on

from google.colab import drive
drive.mount('/content/drive')

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
import random

#associate labels' name to label value in array
classes = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]

#import data and split into training and testing groups
(X_train, y_train), (X_test, y_test) = datasets.cifar10.load_data()
X_train.shape
```

```python
#import the saved model
model = tf.keras.models.load_model('/content/drive/MyDrive/CNNmodel.keras')

#grab a single image to run the model on as an example
img = random.choice(X_train)
img = tf.expand_dims(img, axis=0)
  #fixes this error: Expected shape (None, 32, 32, 3), but input has incompatible shape (32, 32, 3)

#run the model on an image
prediction = model.predict(img)

#print the image and the predicted label
plt.imshow(np.squeeze(img))
plt.xlabel(classes[ tf.argmax(prediction, axis=1).numpy()[0] ])
#this model isn't very good
```
